We decided to measure four features within our metric: most parks, most trees, least tax delinquency, and least condemned properties. The enjoyable aspect describes people's ability to have enjoyable place to go walk, relax, exercise, or play sports. The liveliness aspect describes how much activity is going on in the neighborhood. If there is lower tax deliquency and lower condemnded properties, the neighborhood should be more lively.

The reason we decided on the enjoayable/liveliness metric is because between all of our metrics we have some sort of theme (nature/parks). By combining the metrics to find places with a large amount of green space between trees and parks, it gives a good representation of somewhere we would find enjoyable to live. Likewise, places with tax delinquent and condemned properties impact the liveliness of the area. We worked to find a neighborhood that gave a good balance between number of parks, lots of trees, and few delinquent or condemned properties.

### **Dataset2: Number of Parks in Pittsburgh Neighborhoods**

```python
parksData = pd.read_csv('https://data.wprdc.org/datastore/dump/fa329e3d-89ff-4708-8ddf-81bfedcad11d')
```

```python
parksData.info()
```

```python
parksData.columns
```

```python
parksData
```

#### **Removing Columns**

```python
removeColumns = ['_id', 'id', 'maintenance_responsibility', 'image', 'council_district', 'ward', 'tract', 'public_works_division', 'pli_division', 'police_zone', 'fire_zone', 'latitude', 'longitude']
parksData.drop(columns = removeColumns, inplace = True)
```

#### **Count of Parks, Traffic Island/Medians, Building Grounds, Monument/Memorials, and Open Space Within Each Neighborhood.**

```python
(parksData
    .assign(total = parksData.groupby('neighborhood')['type'].transform('count'),
            count = parksData.groupby(['neighborhood', 'type'])['type'].transform('count'))   
    .drop_duplicates(subset = ['type', 'neighborhood'])
    .sort_values(by = ['total', 'neighborhood', 'count'], ascending = (False, True, False))
    .set_index(['neighborhood', 'type'])
    [['name', 'count', 'total']]
)
```

#### **Filter Non-Park Types**

The reason I am removing every type but the parks is because oftentimes the other types (e.g. monument, memorials, traffic island, medians, open space, and building grounds) are not good representations of parks.
* Traffic islands/medians are not resembalance of parks
* Monuments are not resembalance of parks
* Building grounds often included buildings such as nursing homes or other activity centers
* Open spaces often included plazas (no grass), small get-together spots, etc. Not close enough to the 'parks' type

#### **Count and Name of Parks**

```python
(parksData
    .query('type == "Park"')  
    .assign(count = lambda df: df.groupby('neighborhood')['type'].transform('count'))
    .set_index(['neighborhood', 'name'])
    .sort_values(by = ['count', 'neighborhood', 'name'], ascending = (False, True, True))
    [['type', 'count']]
)
```

#### **Final Parks DataFrame**

```python
parksFinal = parksData
```

```python
parksFinal =( parksFinal
            .query('type == "Park"')  
            .assign(count = lambda df: df.groupby('neighborhood')['type'].transform('count'))
            .set_index(['neighborhood', 'name'])
            .sort_values(by = ['count', 'neighborhood', 'name'], ascending = (False, True, True))
            [['type', 'count']]
            )
```

```python
parksFinal
```

#### **Graph of All Neighborhoods**

```python
(
    parksFinal
        .reset_index()
        .drop_duplicates(subset = 'neighborhood')
        .set_index('neighborhood')
        .drop(columns = ['name', 'type'])
        .sort_values(by = 'count')
        .plot.barh(figsize = (15,15))
    
)
```

#### **Graph of Best 5 Neighborhoods**

```python
(
    parksFinal
        .reset_index()
        .drop_duplicates(subset = 'neighborhood')
        .set_index('neighborhood')
        .drop(columns = ['name', 'type'])
        .sort_values(by = 'count')
        .tail()
        .plot.barh()
    
)
```

### **Dataset4 Number of Condemned and Dead-End Properties Per Neighborhood in Pittsburgh**

```python
import pandas as pd
%matplotlib inline

data = pd.read_csv("condemned_properties.csv")
```

#### C/D-E Prop - Data Gathering: Here, different sets of data are calculated to prove which neighborhood has the least and less serious run-down buildings.

```python
# This dataset just holds the total number of condemned/dead-end properties per neighborhood
nh = data['neighborhood'].value_counts().sort_values(ascending=True)

# This dataset holds the number of failed inspections of those properties per neighborhood
nh_fails = data[data['latest_inspection_result']=='Fail']
nh_fails = nh_fails['neighborhood'].value_counts().sort_values(ascending=True)

# This dataset holds the number of passed inspections of those properties per neighborhood
nh_passes = data[data['latest_inspection_result']=='Pass']
nh_passes = nh_passes['neighborhood'].value_counts().sort_values(ascending=False)

# This dataset holds the average rate of having a passing inspection per neighborhood
nh_avgs = nh_passes / nh
nh_avgs = nh_avgs.fillna(0)
nh_avgs = nh_avgs.sort_values(ascending=False)

# This dataset holds the cumulative inspection scores of each neighborhood
nh_scores = data.groupby('neighborhood')['latest_inspection_score'].sum().sort_values(ascending=False)

# This dataset holds the average inspection scores of each neighborhood
nh_score_avgs = data.groupby('neighborhood')['latest_inspection_score'].mean().sort_values(ascending=False)


# Big Average of all graph data

t = 1.0 - (nh / nh.max())
f = 1.0 - (nh_fails / nh_fails.max())
p = nh_passes / nh_passes.max()
a = nh_avgs / nh_avgs.max()
s = nh_scores / nh_scores.max()
v = nh_score_avgs / nh_score_avgs.max()

data_sum = t + f + p #+ a + s + v
tf = (t + f).fillna(0)
tfp = (tf + p).fillna(0)
tfpa = (tfp + a).fillna(0)
tfpas = (tfpa + s).fillna(0)
tfpasv = (tfpas + v).fillna(0)

total_nh_avgs = tfpasv / 6 #6
total_nh_avgs = total_nh_avgs.sort_values(ascending=False)
```

#### C/D-E Prop - Example Graphs: These graphs condense all the information into just a few neighborhoods to be more easily digestible

```python
# All of the condemned and dead-end properties per neighborhood
nh.plot(
    kind = "bar",
    title="Total Number of Condemned/Dead-End Properties Per Neighborhood in Pittsburgh",
    xlabel="Neighborhood",
    ylabel="Condemned/Dead-End Properties",
    figsize=(20,5)
)
```

```python
# Gets 7 highest amounts of C/D-E Properties
all_check = nh.tail(7)
all_check.plot(
    kind = "bar",
    title="Total Number of Condemned/Dead-End Properties Per Neighborhood in Pittsburgh",
    xlabel="Neighborhood",
    ylabel="Condemned/Dead-End Properties",
    figsize=(5,3)
)
```

```python
# Gets 10 highest amounts of Inspection Passes of C/D-E Properties
passes_check = nh_passes.head(10)
passes_check.plot(
    kind = "bar",
    title="Number of Passing Inspections of Condemned/Dead-End Properties Per Neighborhood in Pittsburgh",
    xlabel="Neighborhood",
    ylabel="Passing Inspections",
    figsize=(5,3)
)
```

```python
# Gets 10 random neighborhood amounts of Inspection Fails of C/D-E Properties
fails_check = nh_fails.sample(10)
fails_check = fails_check.sort_values(ascending=True)
fails_check.plot(
    kind = "bar",
    title="Number of Failed Inspections of Condemned/Dead-End Properties Per Neighborhood in Pittsburgh",
    xlabel="Neighborhood",
    ylabel="Failed Inspections",
    figsize=(5,3)
)
```

```python
# 10 random neighborhood average score for all tests
tot_avgs_check = total_nh_avgs.sample(10) 
tot_avgs_check = tot_avgs_check.sort_values(ascending=False)
tot_avgs_check.plot(
    kind = "bar",
    title="Average Test Score of Condemned/Dead-End Properties Per Neighborhood in Pittsburgh",
    xlabel="Neighborhood",
    ylabel="Average Test Scores",
    figsize=(5,3)
)
```
