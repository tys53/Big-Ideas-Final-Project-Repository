## **Introduction**

We were given the task of coming up with the best neighborhood in Pittsburgh. We used data from the Western Pennsylvania Regional Data Center to determine best neighborhood. Some alternative datasets we considered for our metric included number of ball fields, gyms, air quality, etc. These metrics are similar to some of the ones we choose, but we felt they wete not as representative as we liked. To clarify, the park dataset is a more overarching metric than ball fields or gyms, especially when we wanted this metric to provide people who live in these neighborhood's an outdoor place to relax, exercise, and enjoy life.

We came up with our combined metric after analyzing each member’s choice, and then finding a way to fit them all together. ## **Metric**

**The Metric:** enjoyable/liveliness
We decided to measure four features within our metric: most parks, most trees, least tax delinquency, and least condemned properties. The most parks and most trees are within the enjoyable aspect of our metric, and the least tax deliquency and least condemned properties are within the liveliness aspect of our metric. The enjoyable aspect describes people's ability to have enjoyable place to go walk, relax, exercise, or play sports. The liveliness aspect describes how much activity is going on in the neighborhood. If there is lower tax deliquency and lower condemnded properties, the neighborhood should be more lively. 

The four datasets we chose are City of Pittsburgh Parks, City of Pittsburgh Trees, City of Pittsburgh Property Tax Delinquency, and Condemned and Dead-End Properties. Each dataset provides data on one of the four features.

The reason we decided on the enjoayable/liveliness metric is because between all of our metrics we have some sort of theme (nature/parks). By combining the metrics to find places with a large amount of green space between trees and parks, it gives a good representation of somewhere we would find enjoyable to live. Likewise, places with tax delinquent and condemned properties impact the liveliness of the area. We worked to find a neighborhood that gave a good balance between number of parks, lots of trees, and few delinquent or condemned properties.

### **Dataset2: Number of Parks in Pittsburgh Neighborhoods**

```python
parksData = pd.read_csv('https://data.wprdc.org/datastore/dump/fa329e3d-89ff-4708-8ddf-81bfedcad11d')
```

```python
parksData.info()
```

```python
parksData.columns
```

```python
parksData
```

#### **Removing Columns**

```python
removeColumns = ['_id', 'id', 'maintenance_responsibility', 'image', 'council_district', 'ward', 'tract', 'public_works_division', 'pli_division', 'police_zone', 'fire_zone', 'latitude', 'longitude']
parksData.drop(columns = removeColumns, inplace = True)
```

#### **Count of Parks, Traffic Island/Medians, Building Grounds, Monument/Memorials, and Open Space Within Each Neighborhood.**

```python
(parksData
    .assign(total = parksData.groupby('neighborhood')['type'].transform('count'),
            count = parksData.groupby(['neighborhood', 'type'])['type'].transform('count'))   
    .drop_duplicates(subset = ['type', 'neighborhood'])
    .sort_values(by = ['total', 'neighborhood', 'count'], ascending = (False, True, False))
    .set_index(['neighborhood', 'type'])
    [['name', 'count', 'total']]
)
```

#### **Filter Non-Park Types**

The reason I am removing every type but the parks is because oftentimes the other types (e.g. monument, memorials, traffic island, medians, open space, and building grounds) are not good representations of parks.
* Traffic islands/medians are not resembalance of parks
* Monuments are not resembalance of parks
* Building grounds often included buildings such as nursing homes or other activity centers
* Open spaces often included plazas (no grass), small get-together spots, etc. Not close enough to the 'parks' type

#### **Count and Name of Parks**

```python
(parksData
    .query('type == "Park"')  
    .assign(count = lambda df: df.groupby('neighborhood')['type'].transform('count'))
    .set_index(['neighborhood', 'name'])
    .sort_values(by = ['count', 'neighborhood', 'name'], ascending = (False, True, True))
    [['type', 'count']]
)
```

#### **Final Parks DataFrame**

```python
parksFinal = parksData
```

```python
parksFinal =( parksFinal
            .query('type == "Park"')  
            .assign(count = lambda df: df.groupby('neighborhood')['type'].transform('count'))
            .set_index(['neighborhood', 'name'])
            .sort_values(by = ['count', 'neighborhood', 'name'], ascending = (False, True, True))
            [['type', 'count']]
            )
```

```python
parksFinal
```

#### **Graph of All Neighborhoods**

```python
(
    parksFinal
        .reset_index()
        .drop_duplicates(subset = 'neighborhood')
        .set_index('neighborhood')
        .drop(columns = ['name', 'type'])
        .sort_values(by = 'count')
        .plot.barh(figsize = (15,15))
    
)
```

#### **Graph of Best 5 Neighborhoods**

```python
(
    parksFinal
        .reset_index()
        .drop_duplicates(subset = 'neighborhood')
        .set_index('neighborhood')
        .drop(columns = ['name', 'type'])
        .sort_values(by = 'count')
        .tail()
        .plot.barh()
    
)
```

#### **Graph of Worst 5 Neighborhoods**

```python
(
    parksFinal
        .reset_index()
        .drop_duplicates(subset = 'neighborhood')
        .set_index('neighborhood')
        .drop(columns = ['name', 'type'])
        .sort_values(by = 'count')
        .head()
        .plot.barh()
    
)
```
