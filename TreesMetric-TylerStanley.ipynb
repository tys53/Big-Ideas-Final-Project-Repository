# Total Number of Trees Per Neighborhood 

## My Metric 

WPRDC data on the number of trees in each pittsburgh neighborhood. I love seeing trees and feel like it's impossible to be the "best" neighborhood if you don't have any trees. I took the data from WPRDC and counted how many registered trees were in each neighborhood. I determined this by counting each new line in the dataset as a new tree and pairing that with which neighborhood each of the trees were located in (grouping them together). 

Top 5 by Number of Registered Trees:

### Importing dataset and pandas

```python
import pandas as pd 

df = pd.read_csv("Trees.csv")
df.head()
```

```python
df.columns
```

```python
df = df.sort_values(by="neighborhood")
df.head(10)
```

## Getting the Counts

```python
df = df['neighborhood'].value_counts().sort_values(ascending=False)
df
```

### Plot of All Neighborhoods Based on Number of Registered Trees

```python

df.plot(
    kind = 'bar',
    title = "Number of Registered Trees in Each Pittsburgh Neighborhood",
    xlabel = "Neighborhood",
    ylabel = "Number of Trees",
    figsize = (15,7),

)
plt.show()
```

## Top 5 by Number of Registered Trees
