```python
import pandas as pd

df = pd.read_csv("datasetpgh.csv")
df.head()
```

```python
df.columns
```

```python
grouped = df.groupby("neighborhood")["current_delq_tax"].sum()
# Sort values
sorted_data = grouped.sort_values()

# 5 LOWEST (best)
print("Top 5 BEST (Lowest Tax Delinquency):")
print(sorted_data.head(5))

# 5 HIGHEST (worst)
print("\nTop 5 WORST (Highest Tax Delinquency):")
print(sorted_data.tail(5))
```

```python
lowest5 = sorted_data.head(5)

lowest5.plot(
    kind='bar',
    title="Top 5 Best Neighborhoods (Lowest Tax Delinquency)",
    xlabel="Neighborhood",
    ylabel="Total Delinquent Tax",
    rot=45
)
```

```python
highest5 = sorted_data.tail(5)

highest5.plot(
    kind='bar',
    title="Top 5 Worst Neighborhoods (Highest Tax Delinquency)",
    xlabel="Neighborhood",
    ylabel="Total Delinquent Tax",
    rot=45
)
```
